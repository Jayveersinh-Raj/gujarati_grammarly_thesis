```python
def rouge_1(reference, hypothesis):
    """
    Calculate ROUGE-1 (unigram) scores including recall, precision, and F1 score.

    Args:
    reference (str): The reference (ground truth) sentence.
    hypothesis (str): The generated sentence to compare against the reference.

    Returns:
    dict: A dictionary containing recall, precision, and F1 score.
    """
    ref_tokens = reference.split()
    hyp_tokens = hypothesis.split()

    # Counting the number of unigrams in reference and hypothesis
    ref_count = len(ref_tokens)
    hyp_count = len(hyp_tokens)

    # Counting the number of overlapping unigrams between reference and hypothesis
    overlap_count = len(set(ref_tokens) & set(hyp_tokens))

    # Calculating recall, precision, and F1 score
    recall = overlap_count / ref_count if ref_count > 0 else 0
    precision = overlap_count / hyp_count if hyp_count > 0 else 0
    f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

    return {'recall': recall, 'precision': precision, 'f1': f1}

rouge_1_scores = [rouge_1(ref, hyp) for ref, hyp in zip(dfb['Ground Truth'], dfb['Generated Sentence'])]

# Calculating average scores
avg_recall = sum([score['recall'] for score in rouge_1_scores]) / len(rouge_1_scores)
avg_precision = sum([score['precision'] for score in rouge_1_scores]) / len(rouge_1_scores)
avg_f1 = sum([score['f1'] for score in rouge_1_scores]) / len(rouge_1_scores)

print(f"Average ROUGE-1 Recall: {avg_recall}")
print(f"Average ROUGE-1 Precision: {avg_precision}")
print(f"Average ROUGE-1 F1 Score: {avg_f1}")
```
